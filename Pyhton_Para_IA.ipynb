```python
Lista_de_nomes = [
     ("Carlos Silva"),
    ("Fernanda Santos"),
    ("Roberto Oliveira"),
    ("Patrícia Souza"),
    ("Ricardo Alves")
]

medias_originais = [8.5, 7.9, 9.2, 6.5, 7.0]

# --- Início do Programa Solicitado ---

# Cria uma cópia da lista de médias para alterá-la no loop
medias_trabalho = medias_originais[:]
indice = 0  # Inicializa o índice para o loop while
nota_maxima = 10.0
nota_aprovacao = 6.5

print("--- Processamento de Notas (Ajuste de +1 do trabalho) ---")

# Loop WHILE para percorrer a lista de médias
while indice < len(medias_trabalho):
    nome_completo = f"{Lista_de_nomes[indice]}"
    media_original = medias_originais[indice]

    # 1. Soma +1 à média
    nova_media = media_ajustada = medias_trabalho[indice] + 1

    # 2. Verifica e limita a média a 10.0
    if nova_media > nota_maxima:
        nova_media = nota_maxima

    # Atualiza a média na lista (se houver alteração)
    medias_trabalho[indice] = nova_media

    # 3. Verifica a aprovação/reprovação
    if nova_media > nota_aprovacao:
        status = "APROVADO(A) 🎉"
    else:
        status = "REPROVADO(A) 😔"

    # Imprime o resultado
    print(f"\nAluno(a): {nome_completo}")
    print(f"Média Original: {media_original}")
    print(f"Média Ajustada (+1, máx 10.0): {nova_media}")
    print(f"Status: {status}")

    # Incrementa o índice para mover para o próximo aluno
    indice += 1

print("\n--- Fim do Processamento ---")
print(f"Lista de Médias Finais Ajustadas: {medias_trabalho}")
```

```python
# Dados iniciais (Combinados em um Dicionário para fácil busca)
# {Nome Completo: Média Original}
dados_alunos = {
    "Carlos Silva": 8.5,
    "Fernanda Santos": 7.9,
    "Roberto Oliveira": 9.2,
    "Patrícia Souza": 6.5,
    "Ricardo Alves": 7.0
}

nota_maxima = 10.0
nota_aprovacao = 6.5

print("--- Consulta de Aprovação de Alunos ---")
print("Alunos disponíveis para consulta:", ", ".join(dados_alunos.keys()))
print("-" * 37)

# 1. Receber o nome do usuário
nome_input = input("Digite o nome completo do aluno para verificar (Ex: Carlos Silva): ")

# 2. Normalizar o nome digitado para evitar erros de capitalização
nome_digitado = nome_input.strip().title() # Remove espaços e coloca a primeira letra de cada palavra em maiúscula

# 3. Verificar se o nome está na lista
if nome_digitado in dados_alunos:
    media_original = dados_alunos[nome_digitado]

    # 4. Calcular a nova média (+1, limitado a 10.0)
    nova_media = media_original + 1

    # Aplicar o limite de 10.0
    if nova_media > nota_maxima:
        nova_media = nota_maxima

    # 5. Determinar o status de aprovação
    if nova_media > nota_aprovacao:
        status = "APROVADO(A) 🎉"
    else:
        status = "REPROVADO(A) 😔"

    # 6. Exibir os resultados
    print("\n--- Resultado da Consulta ---")
    print(f"Aluno(a): {nome_digitado}")
    print(f"Média Original: {media_original}")
    print(f"Média Ajustada (+1, máx 10.0): {nova_media:.2f}") # .2f para mostrar 2 casas decimais
    print(f"Status: {status}")
    print("-----------------------------")

else:
    # Caso o nome não seja encontrado
    print(f"\n❌ Erro: O aluno '{nome_digitado}' não foi encontrado na lista.")

# Não foi usado o 'while' aqui pois o objetivo é processar apenas um item
# (o nome digitado) por vez, o que é melhor feito com o 'if/else' de busca.
```

```python
import time # Importa o módulo time

def resumir_email(lista_de_emails):
    for i, email in enumerate(lista_de_emails):
        # 1. Faz a requisição à API
        try:
            resposta = client.models.generate_content(
                model="gemini-flash-latest",
                contents=f"""resuma o corpo do email que eu te mandar em apenas uma linha, me passando o intuito do email: {email}"""
            )
            print(f"--- Resumo do Email {i+1} ---")
            print(resposta.text)
            print("-" * 100)

        except Exception as e:
            # Captura e exibe qualquer erro da API
            print(f"Erro ao processar o Email {i+1}: {e}")
            print("-" * 100)

        # 2. Adiciona o tempo de espera (sleep) de 20 segundos
        # Se for o último email da lista, não precisa esperar
        if i < len(lista_de_emails) - 1:
            print("\n")
            print(f"Aguardando 5 segundos antes do próximo processamento...")
            time.sleep(5) # Pausa a execução por 20 segundos
```

# Lendo e Escrevendo em Arquivos com Python

```python
def resumir_email_corrigido(lista_de_emails):
    lista_resumo = []

    for i, email in enumerate(lista_de_emails):

        try:
            resposta = client.models.generate_content(
                model="gemini-flash-latest",
                contents=f"""resuma o corpo do email que eu te mandar em apenas uma linha, me passando o intuito do email: {email}"""
            )

            cabecalho = f"--- Resumo do Email {i+1} ---"
            resumo_texto = resposta.text.strip() # Remove espaços/quebras de linha extras


            print(cabecalho)
            print(resumo_texto)
            print("-" * 100)


            lista_resumo.append(cabecalho) # Adiciona o cabeçalho
            lista_resumo.append(resumo_texto) # Adiciona o TEXTO do resumo

        except Exception as e:
            # Captura e exibe qualquer erro da API
            erro_msg = f"Erro ao processar o Email {i+1}: {e}"
            print(erro_msg)
            print("-" * 100)
            lista_resumo.append(erro_msg) # Adiciona o erro à lista

            if i < len(lista_de_emails) - 1:
              print("\n")
              time.sleep(10) # Pausa a execução por um tempo

    return lista_resumo


```

```python
lista_resumo = resumir_email_corrigido(emails_gerais)
```
